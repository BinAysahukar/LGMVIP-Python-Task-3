```python
import tkinter as tk
import random

# Function to determine the winner
def determine_winner(player_choice):
    choices = ['rock', 'paper', 'scissors']
    computer_choice = random.choice(choices)
    result_text.set(f"Computer chose {computer_choice}")

    if player_choice == computer_choice:
        result.set("It's a tie!")
    elif (player_choice == 'rock' and computer_choice == 'scissors') or \
         (player_choice == 'paper' and computer_choice == 'rock') or \
         (player_choice == 'scissors' and computer_choice == 'paper'):
        result.set("You win!")
    else:
        result.set("Computer wins!")

# Function to reset the game
def reset_game():
    result.set("")
    result_text.set("")

# Create main window
root = tk.Tk()
root.title("Rock, Paper, Scissors Game")

# Set a larger default window size
root.geometry("600x400")

# Create and place buttons with larger font
button_font = ('Helvetica', 16)
rock_button = tk.Button(root, text="Rock", command=lambda: determine_winner('rock'), font=button_font, width=10, height=2)
rock_button.grid(row=0, column=0, padx=20, pady=20)

paper_button = tk.Button(root, text="Paper", command=lambda: determine_winner('paper'), font=button_font, width=10, height=2)
paper_button.grid(row=0, column=1, padx=20, pady=20)

scissors_button = tk.Button(root, text="Scissors", command=lambda: determine_winner('scissors'), font=button_font, width=10, height=2)
scissors_button.grid(row=0, column=2, padx=20, pady=20)

# Create and place result labels with larger font
result_text = tk.StringVar()
result_text.set("")
result_label = tk.Label(root, textvariable=result_text, font=('Helvetica', 18))
result_label.grid(row=1, column=0, columnspan=3, pady=10)

result = tk.StringVar()
result.set("")
result_label = tk.Label(root, textvariable=result, font=('Helvetica', 24))
result_label.grid(row=2, column=0, columnspan=3, pady=10)

# Create and place reset button with larger font
reset_button = tk.Button(root, text="Play Again", command=reset_game, font=button_font, width=15, height=2)
reset_button.grid(row=3, column=0, columnspan=3, pady=20)

# Start the main loop
root.mainloop()
```
